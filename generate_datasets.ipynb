Read the Pfizer stock prices dataset.

```python
# Convert the 'Date' column in data_pfizer to datetime format
data_pfizer['Date'] = pd.to_datetime(data_pfizer['Date'], format='%Y-%m-%d').dt.date
```

Merge the News dataset pfizer_news with the Stock dataset data_pfizer.

Save the dataset.

```python
merged_df_pfizer.to_csv('/content/drive/MyDrive/Colab Notebooks/[DS] Seminarska/[Code] Models/[Dataset] Generated datasets/pfizer-news-merged.csv', index=False)
```

Group the dataset by columns 'date' and 'newsType'. For each new combination count the number of positive/negative news titles

```python
from collections import Counter

# Group by 'date' and 'newsType', aggregate sentiment scores and count news types
grouped_data = pfizer_news.groupby(['date', 'newsType']).agg({
    'sentimentScore': lambda x: Counter(x),
    'newsType': 'count'
}).rename(columns={'newsType': 'newsType_count'}).reset_index()

print(grouped_data)
```

```python
# Initialize new columns for percentages
grouped_data['negative_percentage'] = 0.0
grouped_data['positive_percentage'] = 0.0

# Iterate through the rows to calculate the percentages
for i, row in grouped_data.iterrows():
    sentiment_dict = row['sentimentScore']

    # Get the counts of NEGATIVE and POSITIVE sentiments
    negative_count = sentiment_dict.get('NEGATIVE', 0)
    positive_count = sentiment_dict.get('POSITIVE', 0)
    total_count = row['newsType_count']

    # Calculate the percentages and assign to the new columns
    if total_count > 0:
        grouped_data.at[i, 'negative_percentage'] = (negative_count / total_count) * 100
        grouped_data.at[i, 'positive_percentage'] = (positive_count / total_count) * 100

# Display the updated DataFrame with the new columns
print(grouped_data[['date', 'newsType', 'negative_percentage', 'positive_percentage']])
```

```python
grouped_data.to_csv("/content/drive/MyDrive/Colab Notebooks/[DS] Seminarska/[Code] Models/[Dataset] Generated datasets/pfizer-grouped.csv", index=False)
```

Now, group the previous dataset by 'date' and count the percentage of positive/negative news titles for each news type.

```python
# Pivot the DataFrame to get each newsType as separate columns for negative and positive percentages
pivot_df_pfizer = grouped_data.pivot(index='date', columns='newsType', values=['negative_percentage', 'positive_percentage'])

# Flatten the multi-level column names
pivot_df_pfizer.columns = [f"{col[1]}_{col[0]}" for col in pivot_df_pfizer.columns]

# Reset the index to bring 'Date' back as a column
pivot_df_pfizer = pivot_df_pfizer.reset_index()

# Display the resulting DataFrame
print(pivot_df_pfizer.head())
```

```python
pivot_df_pfizer.head(10)
```

Now, we will merge the data_pfizer with pivot_df_pfizer.

```python
pfizer_final = pd.merge(data_pfizer, pivot_df_pfizer, on=['date'])
```

```python
pfizer_final.head(10)
```

```python
pfizer_final.describe()
```

```python
pfizer_final.info()
```

```python
pfizer_final.to_csv("/content/drive/MyDrive/Colab Notebooks/[DS] Seminarska/[Code] Models/[Dataset] Generated datasets/pfizer-final.csv", index=False)
```
