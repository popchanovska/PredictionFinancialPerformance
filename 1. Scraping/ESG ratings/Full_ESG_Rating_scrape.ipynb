```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def get_esg_rating(ticker):
    url = f"https://finance.yahoo.com/quote/{ticker}/sustainability/"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
    }
    response = requests.get(url, headers=headers)

    if response.status_code != 200:
        return None

    soup = BeautifulSoup(response.text, 'html.parser')
    try:
        rating_div = soup.find('div', class_='scoreRank')
        if rating_div:
            rating = rating_div.find('h4', class_='border yf-y3c2sq').text.strip()
            return rating
    except AttributeError:
        return None

    return None
```

```python
input_csv = '/content/drive/MyDrive/vnp-project/data/results_esg_index.csv'
tickers_df = pd.read_csv(input_csv)
tickers = tickers_df[['ticker', 'company']]
```

```python
results = []
for _, row in tickers.iterrows():
    ticker = row['ticker']
    company_name = row['company']
    rating = get_esg_rating(ticker)
    results.append({'Ticker': ticker, 'Company Name': company_name, 'ESG Rating': rating})
```

```python
results_df = pd.DataFrame(results)
output_csv = '/content/drive/MyDrive/vnp-project/data/esg_ratings.csv'
results_df.to_csv(output_csv, index=False)
```
