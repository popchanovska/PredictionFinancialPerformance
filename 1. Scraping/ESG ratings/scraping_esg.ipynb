```python
file_path = '/content/drive/MyDrive/Colab Notebooks/[DS] Seminarska/companies-tickers.xlsx'

df = pd.read_excel(file_path)

print(df.columns)

ticker_array = df['ticker_code'].tolist()
```

```python
ticker_array
```

```python
import yfinance as yf
import pandas as pd
import os
```

```python
esg_index = []
```

```python
for ticker in ticker_array:
  try:
    ticker_value = ticker
    ticker_y = yf.Ticker(ticker_value)
    esg_data = pd.DataFrame.transpose(ticker_y.sustainability)
    esg_data['company_ticker'] = str(ticker_y.ticker)

    if 'environmentScore' in esg_data.columns and 'governanceScore' in esg_data.columns and 'socialScore' in esg_data.columns:
            esg_index.append({
                'ticker': ticker,
                'environmentScore': esg_data['environmentScore'].values[0],
                'socialScore': esg_data['socialScore'].values[0],
                'governanceScore': esg_data['governanceScore'].values[0]
            })

  except IndexError:
    pass
```

```python
esg_index
```

```python
esg_index_df = pd.DataFrame(esg_index)
```
