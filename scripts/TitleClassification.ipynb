```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification", model="MoritzLaurer/mDeBERTa-v3-base-mnli-xnli")
```

```python
import pandas as pd
```

```python
news = pd.read_csv("/content/drive/MyDrive/vnp-project/ana/news.csv", index_col=0)
```

```python
labels = ["environment", "social", "government", "neutral"]
```

```python
data = []
```

```python
import numpy as np
```

```python
for _, row in news.iterrows():
  title = row["title"]
  company = row["company"]
  output = classifier(title, labels, multi_label=False)
  label = output["labels"][np.argmax(output["scores"])]
  probability = np.max(output["scores"])
  data.append(
      {
          "company": company,
          "title": title,
          "link": row["link"],
          "publish_date": row["publish_date"],
          "label": label,
          "probability": probability
      }
  )
```

```python
data
```

```python
zero_shot_classification = pd.DataFrame(data)
zero_shot_classification.to_csv("/content/drive/MyDrive/vnp-project/ana/news_titles_classification_new1.csv", index=False)
```
