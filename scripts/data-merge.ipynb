```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/vnp-project/data/news_data.csv")
```

```python
df["publish_date"] = pd.to_datetime(df["publish_date"])
df['date'] = df['publish_date'].dt.strftime('%Y-%m-%d')
df = df.drop(columns=['publish_date'])
df
```

```python
df.rename(columns={'date': 'Date'}, inplace=True)
```

```python
stocks = pd.read_csv("/content/drive/MyDrive/vnp-project/data/stocks.csv")
```

```python
esg = pd.read_csv("/content/drive/MyDrive/vnp-project/data/results_esg_index.csv")
```

```python
data_temp = pd.merge(df,esg,on="company")
data_temp
```

```python
data = pd.merge(data_temp,stocks,on=["company","Date"], how="left")
data
```

```python
data = data.dropna()
data
```

```python
data.to_csv("/content/drive/MyDrive/vnp-project/data/data.csv")
```
