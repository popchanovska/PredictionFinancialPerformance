```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification", model="MoritzLaurer/mDeBERTa-v3-base-mnli-xnli")
```

```python
import pandas as pd
```

```python
news_with_content = pd.read_csv("/content/drive/MyDrive/vnp-project/raw-news/csv_files/news_with_content.csv", index_col=0)
news_without_content = pd.read_csv("/content/drive/MyDrive/vnp-project/raw-news/csv_files/news_without_content.csv", index_col=0)
```

```python
labels = ["environment", "social", "government", "neutral"]
```

```python
data = []
```

```python
import numpy as np
```

```python
for _, row in news_without_content.iterrows():
  title = row["title"]
  company = row["company"]
  output = classifier(title, labels, multi_label=False)
  label = output["labels"][np.argmax(output["scores"])]
  probability = np.max(output["scores"])
  data.append(
      {
          "company": company,
          "title": title,
          "label": label,
          "probability": probability
      }
  )
```

```python
for _, row in news_with_content.iterrows():
  title = row["title"]
  company = row["company"]
  output = classifier(title, labels, multi_label=False)
  label = output["labels"][np.argmax(output["scores"])]
  probability = np.max(output["scores"])
  data.append(
      {
          "company": company,
          "title": title,
          "label": label,
          "probability": probability
      }
  )
```

```python
zero_shot_classification = pd.DataFrame(data)
zero_shot_classification.to_csv("/content/drive/MyDrive/vnp-project/zero-shot/news_titles_classification.csv", index=False)
```
